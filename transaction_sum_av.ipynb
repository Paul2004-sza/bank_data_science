```python
import pandas as pd
import plotly.express as px

class TransactionAnalyzer:
    def __init__(self, filepath):
        self.filepath = filepath
        self.df = None
        self.account_stats = None

    def load_data(self):
        self.df = pd.read_csv(self.filepath)
        print(f"Data loaded successfully from {self.filepath}")

    def calculate_account_statistics(self):
        self.account_stats = self.df.groupby('account_id')['amount'].agg(
            ['mean', 'count', 'sum', 'min', 'max', 'std']
        ).reset_index()

        self.account_stats.columns = [
            'account_id',
            'average_transaction',
            'transaction_count',
            'total_amount',
            'min_transaction',
            'max_transaction',
            'transaction_std_dev'
        ]
        print("Account statistics calculated successfully.")

    def display_statistics(self):
        if self.account_stats is not None:
            print("Transaction Statistics per Account:")
            display(self.account_stats.head())
        else:
            print("No statistics available. Please run calculate_account_statistics() first.")

    def save_statistics(self, output_path):
        if self.account_stats is not None:
            self.account_stats.to_csv(output_path, index=False)
            print(f"Statistics saved to '{output_path}'")
        else:
            print("No statistics to save. Please run calculate_account_statistics() first.")

    def plot_histogram(self):
        if self.account_stats is not None:
            fig = px.histogram(
                self.account_stats,
                x='average_transaction',
                nbins=30,
                marginal='box',  # Adds a box plot above
                title='Distribution of Average Transaction Amounts per Account',
                labels={
                    'average_transaction': 'Average Transaction Amount'
                },
                color_discrete_sequence=['#00CC96']
            )

            fig.update_layout(
                title_font_size=24,
                xaxis_title_font_size=18,
                yaxis_title_font_size=18,
                bargap=0.2,
                template='plotly_white',
                hoverlabel_font_size=14
            )

            fig.show()
        else:
            print("No data to plot. Please run calculate_account_statistics() first.")

    def plot_scatter(self):
        if self.account_stats is not None:
            fig2 = px.scatter(
                self.account_stats,
                x='account_id',
                y='average_transaction',
                size='transaction_count',
                color='total_amount',
                title='Account-wise Average Transaction Amounts',
                labels={
                    'account_id': 'Account ID',
                    'average_transaction': 'Average Transaction Amount',
                    'transaction_count': 'Number of Transactions',
                    'total_amount': 'Total Transaction Amount'
                },
                color_continuous_scale='Viridis'
            )

            fig2.update_layout(
                title_font_size=24,
                xaxis_title_font_size=18,
                yaxis_title_font_size=18,
                template='plotly_white',
                hoverlabel_font_size=14
            )

            fig2.show()
        else:
            print("No data to plot. Please run calculate_account_statistics() first.")

analyzer = TransactionAnalyzer('C:/Users/Sut Zaw Aung/StockMarketDB/pythonProjectBank/data/row/c_transactions_cleaned.csv')

analyzer.load_data()
analyzer.calculate_account_statistics()
analyzer.display_statistics()
analyzer.save_statistics('transaction_statistics_per_account.csv')
analyzer.plot_histogram()
analyzer.plot_scatter()
```
