```python
import pandas as pd

accounts = pd.read_csv('C:/Users/Sut Zaw Aung/StockMarketDB/pythonProjectBank/data/row/c_accounts.csv')
customers = pd.read_csv('C:/Users/Sut Zaw Aung/StockMarketDB/pythonProjectBank/data/row/b_customers.csv')

merged = pd.merge(accounts, customers, on='customer_id')

status_count = merged.groupby(['region', 'status']).size().reset_index(name='status_count')
print("Status Count by Region")
display(status_count)

status_balance_region = merged.groupby(['region', 'status'])['balance'].sum().reset_index()
status_balance_region = status_balance_region.sort_values(by='balance', ascending=False)
print("Balance by Region and Status")
display(status_balance_region)

region_balance = merged.groupby('region')['balance'].sum().reset_index()
region_balance = region_balance.sort_values(by='balance', ascending=False)
print("Total Balance by Region")
display(region_balance)
```
